## (h) let's explore graduation rates

````python
# Use Seaborn style
sns.set(style="whitegrid")

# Create a 3x3 grid of subplots
fig, axes = plt.subplots(3, 3, figsize=(18, 15))

# Plot 1: Graduation Rate by Private/Public
sns.boxplot(x='Private', y='Grad.Rate', data=college, ax=axes[0, 0])
axes[0, 0].set_title('Graduation Rate by Private/Public')

# Plot 2: Top10perc vs Graduation Rate
sns.scatterplot(x='Top10perc', y='Grad.Rate', data=college, ax=axes[0, 1])
axes[0, 1].set_title('Top 10% of HS Class vs Graduation Rate')

# Plot 3: Top25perc vs Graduation Rate
sns.scatterplot(x='Top25perc', y='Grad.Rate', data=college, ax=axes[0, 2])
axes[0, 2].set_title('Top 25% of HS Class vs Graduation Rate')

# Plot 4: Outstate Tuition vs Graduation Rate
sns.scatterplot(x='Outstate', y='Grad.Rate', data=college, ax=axes[1, 0])
axes[1, 0].set_title('Out-of-State Tuition vs Graduation Rate')

# Plot 5: Room and Board vs Graduation Rate
sns.scatterplot(x='Room.Board', y='Grad.Rate', data=college, ax=axes[1, 1])
axes[1, 1].set_title('Room & Board Cost vs Graduation Rate')

# Plot 6: Faculty PhD % vs Graduation Rate
sns.scatterplot(x='PhD', y='Grad.Rate', data=college, ax=axes[1, 2])
axes[1, 2].set_title('Faculty PhD % vs Graduation Rate')

# Plot 7: Alumni Donation % vs Graduation Rate
sns.scatterplot(x='perc.alumni', y='Grad.Rate', data=college, ax=axes[2, 0])
axes[2, 0].set_title('Alumni Donation % vs Graduation Rate')

# Plot 8: Instructional Expenditure vs Graduation Rate
sns.scatterplot(x='Expend', y='Grad.Rate', data=college, ax=axes[2, 1])
axes[2, 1].set_title('Instructional Expenditure vs Graduation Rate')

# Plot 9: Student-Faculty Ratio vs Graduation Rate
sns.scatterplot(x='S.F.Ratio', y='Grad.Rate', data=college, ax=axes[2, 2])
axes[2, 2].set_title('Student-Faculty Ratio vs Graduation Rate')

# Adjust layout
plt.tight_layout()
plt.suptitle('Factors Affecting Graduation Rate', fontsize=20, y=1.02)
plt.show()
````

# Key Observations
## Private vs Public:
Private colleges generally have higher graduation rates than public ones.

## Top10perc / Top25perc:
A strong positive correlation: the more students from the top of their high school class, the higher the graduation rate.

## Outstate Tuition & Room.Board:

Mild positive relationship: higher-cost institutions may offer more resources, slightly boosting graduation rates.

## Faculty with PhDs:

Slight positive correlation; better-qualified faculty might improve student outcomes.

## Alumni Donations:

Clear upward trend: higher alumni engagement correlates with better graduation rates.

## Instructional Expenditure:

Institutions that spend more per student tend to have higher graduation rates.

## Student-Faculty Ratio:

Inverse relationship: lower ratios (more faculty per student) lead to better graduation outcomes.
